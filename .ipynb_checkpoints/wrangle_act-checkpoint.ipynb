### Issue #1:  `twitter_archive` table
Inadequate type of some variables:tweet_id is an integer, retweeted_status_id and retweeted_status_user_id are float, timestamp and retweeted_status_timestamp are objects

#### Define:
- Change the type of:
     - tweet_id, in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id and retweeted_status_user_id to string
     - timestamp and retweeted_status_timestamp to datetimes 

#### Code

```python
#change the type of tweet_id, retweeted_status_id and retweeted_user_id 
# from integer to string by using .astype() and .map() functions
twarchive_clean['tweet_id']=twarchive_clean['tweet_id'].astype(str)
twarchive_clean['in_reply_to_status_id']=twarchive_clean['in_reply_to_status_id'].astype(str)
twarchive_clean['in_reply_to_user_id']=twarchive_clean['in_reply_to_user_id'].astype(str)
twarchive_clean['retweeted_status_id']=twarchive_clean['retweeted_status_id'].map(str)
twarchive_clean['retweeted_status_user_id']=twarchive_clean['retweeted_status_user_id'].map(str)
```

```python
# Change the type of timestamp and retweeted_status_timestamp to datetimes using the pd.to_datetime function
twarchive_clean['timestamp']= pd.to_datetime(twarchive_clean['timestamp'])
twarchive_clean['retweeted_status_timestamp']= pd.to_datetime(twarchive_clean['retweeted_status_timestamp'])
```

#### Test

```python
# check
twarchive_clean.info() #types of concerned data are changed
```

### Issue #2: `twitter_archive` table
None instead of NaN in doggo, floofer, pupper,and puppo columns

#### Define
Replace None in the columns doggo, floofer, pupper and puppo by NaN using the .replace function

#### Code

```python
twarchive_clean['doggo'].replace('None', np.nan, inplace=True)
twarchive_clean['floofer'].replace('None', np.nan, inplace=True)
twarchive_clean['pupper'].replace('None', np.nan, inplace=True)
twarchive_clean['puppo'].replace('None', np.nan, inplace=True)
```

### Issue #3: `twitter_archive` table
745 None dog's name instead of NaN, inproper names like:"a", "the", "an" and "his": name starting with small letters

#### Define


Create a definition to replace all inappropriate name that starting with small letter with None
Replace all None in dog name with NaN by using the .replace function 

#### Code

```python
list(twarchive_clean.name)
# Define function to replace incorrect names to 'None'
def inappropriate_name(row):
    if (row['name'][0].islower() == False) and (row['name'][0]!= 'None'): 
        return row['name']
    else:
        return 'None'

#call function and apply by column in the twitter archive table
twarchive_clean['name']=twarchive_clean.apply(inappropriate_name, axis=1)      
```

#### Test

```python
twarchive_clean.name 
```

```python
# replace all None dog's name to NaN
twarchive_clean['name'].replace('None', np.nan, inplace=True)
```

### Tidiness issues
1. (doggo, floofer, pupper and puppo) include the same variable "stages of dogs" and can be merged in one column

2. Three tables with different lengths that can be merged in one dataframe.

#### Define

Include the four stages of dog ages as one variable named dog_stages using the extract module.

```python
twarchive_clean.info()
```

#### Code

```python
#extract the data from the type of dogs and merge it in one column named dog_stages.
twarchive_clean['dog_stages']=twarchive_clean['text'].str.extract('(doggo|floofer|pupper|puppo)', expand=True)
twarchive_clean=twarchive_clean.drop(columns=['doggo', 'floofer', 'pupper', 'puppo'],axis=1)
```

#### Test

```python
twarchive_clean.info()
```

```python
twarchive_clean.dog_stages.value_counts()
```

#### Define

Merge the three table of dog data in one dataframe by using the merge function and inner parameter

#### Code

```python
#merge the three table based on the tweet_id 
twitter_archive_master=twarchive_clean.merge(imagepred_clean.merge(favorite_clean, how='inner', on='tweet_id'),how='inner', on='tweet_id')
```

#### Test

```python
twitter_archive_master.info()
```

### Insights:
1. How is the correlation between favorite count and retweet count?

2. What is the most used name for dogs?

3. What is the highly rated dog stage?

4. What is the most truly identified dog breed by the neural network according to the image's dog?

```python
#calculate the mean of the rating given to the dogs 
twitter_archive_master.rating_numerator.mean()
```

```python
#distribution of rating of dogs 
twitter_archive_master.rating_numerator.hist()
plt.title('Rating Distribution')
plt.xlabel('Rate')
plt.ylabel('Number of dog');
```

```python
# the rating of dog according to stages
rating_dog_stages=twitter_archive_master.groupby(['dog_stages'])['rating_numerator'].mean().sort_values(ascending=False)
rating_dog_stages
```

```python
#create a bar chart regarding the rating of dogs based on their stages
locations=[1,2,3,4]
heights=rating_dog_stages
labels=['floofer','puppo','doggo','pupper']
plt.bar(locations, heights, tick_label=labels, width=0.6)
plt.title('Rating of dogs according to their stages')
plt.xlabel('Dog stages')
plt.ylabel('Dog rates');
```

```python
# Do the tweet who have the highest favorite_count retweeted more than lowest ones? 
# scatter plot showing correlation between favorite_count and retweet_count 
sns.regplot(data=twitter_archive_master, x= 'favorite_count', y='retweet_count', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20});
```

```python
#plot the 10 most common used name in bar chart
plt.figure(figsize=(15,8))
plt.title("10 Most common dog name used", size=30)
most_used_name=twitter_archive_master.name.value_counts().sort_values(ascending=False).head(10)
most_used_name.plot(kind='barh', fontsize=15, color='blue', alpha=0.6);
```

### Visualization

```python
#select the #1 prediction which is true 
right_prediction=twitter_archive_master.p1_dog==True
# select the #1 prediction which is wrong
wrong_prediction=twitter_archive_master.p1_dog==False
```
